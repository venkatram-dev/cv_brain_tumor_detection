```python
import matplotlib.pyplot as plt
import cv2
```

```python
yes_folder_location = "../data/yes"
```

```python
yes_file_name = f"{yes_folder_location}/y0.jpg"
yes_img_sample = cv2.imread(yes_file_name)
```

```python
yes_img_sample.shape
```

```python
plt.imshow(yes_img_sample)
```

```python
yes_img_sample_converted_color = cv2.cvtColor(yes_img_sample,cv2.COLOR_BGR2RGB)
plt.imshow(yes_img_sample_converted_color)
```

```python
yes_img_sample_converted_gray = cv2.cvtColor(yes_img_sample,cv2.COLOR_BGR2GRAY)
print (yes_img_sample_converted_gray.shape)
plt.imshow(yes_img_sample_converted_gray)
```

```python
yes_img_sample_converted_gray_rev = cv2.cvtColor(yes_img_sample,cv2.COLOR_RGB2GRAY)
print (yes_img_sample_converted_gray_rev.shape)
plt.imshow(yes_img_sample_converted_gray_rev)
```

```python
no_folder_location = "../data/no"
no_file_name = f"{no_folder_location}/no0.jpg"
no_img_sample = cv2.imread(no_file_name)
print (no_img_sample.shape)
plt.imshow(no_img_sample)
```

```python
desired_image_shape = (250,250)
reshaped_yes_image = cv2.resize(yes_img_sample,desired_image_shape)
print (reshaped_yes_image.shape)
plt.imshow(reshaped_yes_image)
```

```python
reshaped_no_image = cv2.resize(no_img_sample,desired_image_shape)
print (reshaped_no_image.shape)
plt.imshow(reshaped_no_image)
```
